```python
import torch
from torchvision import datasets, transforms
import matplotlib.pyplot as plt
```

```python
transform = transforms.Compose([
    transforms.ToTensor(),
])
```

```python
train_dataset = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
test_dataset = datasets.MNIST(root='./data', train=False, download=True, transform=transform)
```

```python
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=10, shuffle=True)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=1, shuffle=False)
```

```python
unique_samples = {}

for images, labels in train_loader:
    label = labels[0].item()
    if label not in unique_samples:
        unique_samples[label] = images.squeeze().numpy()
        if len(unique_samples) == 10:
            break
```

## بخش الف

```python
#train
print("تعداد نمونه‌های داده‌های آموزش:", len(train_dataset))
print("ابعاد داده‌های آموزش:", train_dataset.data.shape)
# test
print("تعداد نمونه‌های داده‌های آزمون:", len(test_dataset))
print("ابعاد داده‌های آزمون:", test_dataset.data.shape)
```

## بخش ب

```python
num_samples = 10
train_loader_list = list(train_loader)
images, labels = train_loader_list[0]
plt.figure(figsize=(15, 10))

for j in range(images.size(0)):
        plt.subplot(1, num_samples, j + 1)
        plt.imshow(images[j].squeeze().numpy(), cmap='gray')
        plt.title('Label: ' + str(labels[j].item()))
        plt.axis('off')
plt.show()
```

## بخش ج

```python
# محاسبه تعداد نمونه‌های هر کلاس برای داده‌های آموزش
train_class_counts = [0] * 10
for _, labels in train_loader:
    for label in labels:
        train_class_counts[label.item()] += 1

# محاسبه تعداد نمونه‌های هر کلاس برای داده‌های آزمون
test_class_counts = [0] * 10
for _, labels in test_loader:
    for label in labels:
        test_class_counts[label.item()] += 1

# رسم نمودار هیستوگرام
plt.figure(figsize=(10, 5))
plt.bar(range(10), train_class_counts, alpha=0.7, color='blue', label='Train')
plt.bar(range(10), test_class_counts, alpha=0.7, color='red', label='Test')
plt.xlabel('Classes')
plt.ylabel('Count')
plt.title('Count of Training and Testing Samples')
plt.xticks(range(10))
plt.legend()
plt.show()
```

## بخش د

```python
train_images = train_dataset.data.float()  # تبدیل تصاویر به نوع داده‌های پیکسل‌ها به float
min_val = train_images.min().item()
max_val = train_images.max().item()
normalized_images = (train_images - min_val) / (max_val - min_val)

# چاپ مقادیر مینیمم و ماکسیمم برای اطمینان از صحت عملیات نرمال‌سازی
print("Min:", normalized_images.min().item())
print("Max:", normalized_images.max().item())
```
